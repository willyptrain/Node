```
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

data = pd.DataFrame()
apps = ["chrome", "youtube", "snapchat", "gmail", "messenger"]
usage = [25, 18, 16, 10, 8]
data['apps'] = apps
data['usage'] = usage
plt.figure(figsize=(6,6))
plt.pie(usage, labels=data['apps'], shadow=True, autopct='%1.0f%%')
plt.title("phone usage")
plt.show()
```

```
sleep_data = pd.read_csv('nodea_sleep_data.csv')
```

```
sleep_data.head()
```

```
def mood_to_color(row):
  if(row == ":)"):
    return "green"
  else:
    return "red"


sleep_data['color'] = sleep_data['Wake up'].apply(mood_to_color)
plt.scatter(sleep_data['Time in bed'], sleep_data['Sleep quality'], color=sleep_data['color'])
plt.legend(sleep_data['color'])
plt.show()
```
